```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/Practice dataset 5.csv')
df = data
```

```python
df
```

```python
sns.countplot(x='gender',data=df,hue='diabetes',palette="Set2")
plt.show()
```

```python
from seaborn.categorical import countplot
sns.countplot(y='gender',data=df,hue='smoking_history',palette="Set1")
plt.show()
```

```python
correlation_matrix = data.corr()
print(correlation_matrix)
```

```python
sns.heatmap(data.corr(),annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.figure(figsize=(5, 5))
plt.show()
```
