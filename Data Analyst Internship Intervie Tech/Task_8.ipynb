```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/Practice dataset 5.csv')
df = data
```

```python
df
```

```python
correlation_matrix = data.corr()
print(correlation_matrix)
```

```python
plt.hist(df.gender, bins=3, edgecolor='k')
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
plt.scatter(x=df['gender'], y=df['smoking_history'], c='g', marker='x', label='Data')
plt.xlabel('Gender')
plt.ylabel('smoking_history')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```

```python
plt.bar(df.gender,df.age, color='skyblue')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Vertical Bar Chart')
plt.show()
```

```python
plt.barh(df.gender,df.age, color='skyblue')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Horizontal Bar Chart')
plt.show()
```

```python
plt.boxplot(df.age)
plt.ylabel('Age')
plt.title('Box Plot')
plt.show()
```

```python
sns.boxplot(y=df.HbA1c_level,width=0.2)
plt.ylabel('HbA1c_level')
plt.title('Box Plot')
plt.show()
```

```python
plt.plot(df.gender,df.age,linestyle='-',marker='x',markerfacecolor='blue', color='lightblue', label='Data')
plt.ylabel('Heart_disease rate')
plt.xlabel('Gender')
plt.title('Line Chart')
plt.legend()
plt.show()
```
